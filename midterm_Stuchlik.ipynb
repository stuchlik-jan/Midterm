### Part 1

```python
# Importing all necessary libraries

import requests
from bs4 import BeautifulSoup
from time import sleep
from tqdm import tqdm
import re
import pandas as pd
import json
```

```python
# Defining a function for web scraping

def getSoup(link):
    sleep(0.1) # being nice to the website
    r = requests.get(link)
    r.encoding = 'UTF-8'
    return BeautifulSoup(r.text,'lxml')

link = 'https://en.wikipedia.org/wiki/FTSE_100_Index'
soup = getSoup(link)

# Extracting the desired table

contents = soup.find('table', {'id':'constituents'}).findAll('tr')
print(contents)

tickers = []
industries = []

# Looping over all rows of the table

for content in contents:
    ticker = content.td[1].text
    industry = content.td[2].text
    tickers.append(ticker)
    industries.append(industry)
    
print(tickers)
print(industries)

# Creating a dataframe

df = pd.DataFrame({'ticker': tickers,
                  'industry': industries
                  })
```
